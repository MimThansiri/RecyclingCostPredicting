# SET INSTANCE FOR TABLE 1

```python
steel = soup.find_all('table')[1]
print(steel)
```

# FIND TITLE

```python
steel_title = steel.find_all('th')
print(steel_title)
```

```python
steel_table_title = [title.text for title in steel_title]
print(steel_table_title)
```

# Extract the Main Title

```python
main_title = steel_title[0].text.strip()
print(main_title)
```

# Subcategory of Main Title

```python
subcategories = steel_title[1].text.strip()
price = steel_title[2].text.strip()

print(subcategories)
print(price)
```

```python
data = {
    (main_title, subcategories): [90, 85, 88],
    (main_title, price): [75, 80, 85]
}

df = pd.DataFrame(data)

# Rename the index to give it a main title
df.index = ['Student 1', 'Student 2', 'Student 3']

# Display DataFrame with gridlines
styled_df = df.style.set_table_styles([{
    'selector': 'th',
    'props': [('border', '1px solid black')]
}, {
    'selector': 'td',
    'props': [('border', '1px solid black')]
}])

print(styled_df)
```
